```python
import pandas as pd
```

```python
data1 = pd.read_csv("JC-201901-citibike-tripdata.csv", error_bad_lines=False)
data2 = pd.read_csv("JC-201902-citibike-tripdata.csv", error_bad_lines=False)
data3 = pd.read_csv("JC-201903-citibike-tripdata.csv", error_bad_lines=False)
data4 = pd.read_csv("JC-201904-citibike-tripdata.csv", error_bad_lines=False)
data5 = pd.read_csv("JC-201905-citibike-tripdata.csv", error_bad_lines=False)
data6 = pd.read_csv("JC-201906-citibike-tripdata.csv", error_bad_lines=False)
data7 = pd.read_csv("JC-201907-citibike-tripdata.csv", error_bad_lines=False)
data8 = pd.read_csv("JC-201908-citibike-tripdata.csv", error_bad_lines=False)
data9 = pd.read_csv("JC-201909-citibike-tripdata.csv", error_bad_lines=False)
data10 = pd.read_csv("JC-201910-citibike-tripdata.csv", error_bad_lines=False)
data11 = pd.read_csv("JC-201911-citibike-tripdata.csv", error_bad_lines=False)
data12 = pd.read_csv("JC-201912-citibike-tripdata.csv", error_bad_lines=False)
```

```python
tripdata_2019 = [data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12]
citibike_2019 = pd.concat(tripdata_2019)
citibike_2019.head()
```

```python
citibike_2019.gender = citibike_2019.gender.astype(str)
```

```python
citibike_2019['gender'].replace({'1': 'male', '2': 'female'}, inplace=True)
```

```python
citibike_2019.to_csv("2019_citibike_tripdata_cleaned.csv", index=False)
```
